```python
from sqlalchemy import create_engine
import pandas as pd

# Substitua com seus dados reais
usuario = "root"
senha = "8967Clarah"
host = "localhost"
porta = 3306
banco = "focos_db"

# Criar engine
engine = create_engine(f"mysql+pymysql://root:8967Clarah@localhost:3306/focos_db")

# Carregar dados
df = pd.read_sql("SELECT * FROM focos_norte", engine)

# Converter data_pas para datetime e extrair o ano
df['data_pas'] = pd.to_datetime(df['data_pas'], errors='coerce')
df['ano'] = df['data_pas'].dt.year
```

```python
#Total de focos por estado e bioma
total_focos = (
    df.groupby(['estado', 'bioma', 'ano'])
    .size()
    .reset_index(name='total_focos')
    .sort_values(by='total_focos', ascending=False)
)

print("Total de focos por estado, bioma e ano:")
print(total_focos)
```

```python
#Evolução anual por estado
evolucao_anual = (
    df.groupby(['estado', 'bioma', 'ano'])
    .size()
    .reset_index(name='total')
    .sort_values(by=['estado', 'bioma', 'ano'])
)

print("\nEvolução anual por estado e bioma:")
print(evolucao_anual)
```

```python
#Estados com mais de 1000 focos
focos_estado_bioma = (
    df.groupby(['estado', 'bioma'])
    .size()
    .reset_index(name='total')
)

mais_de_1000 = focos_estado_bioma[focos_estado_bioma['total'] > 1000].sort_values(by='total', ascending=False)

print("\nEstados com mais de 1000 focos por bioma:")
print(mais_de_1000)
```

```python
#Comparar focos entre três anos (2022, 2023, 2024)
anos_filtrados = df[df['ano'].isin([2022, 2023, 2024])]

comparacao_3anos = (
    anos_filtrados.groupby(['estado', 'bioma', 'ano'])
    .size()
    .reset_index(name='total')
    .sort_values(by=['estado', 'bioma', 'ano'])
)

print("\nComparação de focos entre 2022, 2023 e 2024:")
print(comparacao_3anos)
```

```python
#Comparar focos entre dois anos (mesmo resultado da anterior)
comparacao_2anos = (
    df[df['ano'].isin([2023, 2024])]
    .groupby(['municipio', 'ano'])
    .size()
    .reset_index(name='total')
    .sort_values(by='total', ascending=False)
)

print("\nComparação de focos entre 2023 e 2024 por município:")
print(comparacao_2anos)
```

```python
#Top 10 municípios com mais focos
top_municipios = (
    df.groupby(['municipio', 'ano'])
    .size()
    .reset_index(name='total_focos')
    .sort_values(by='total_focos', ascending=False)
    .head(10)
)

print("\nTop 10 municípios com mais focos por ano:")
print(top_municipios)
```

```python
def insertion_sort_bioma(data):
    data = data.copy()
    comparacoes = 0
    trocas = 0

    for i in range(1, len(data)):
        chave = data[i]
        j = i - 1

        while j >= 0 and data[j]['bioma'] is not None and chave['bioma'] is not None and data[j]['bioma'] > chave['bioma']:
            comparacoes += 1
            data[j + 1] = data[j]
            trocas += 1
            j -= 1

        data[j + 1] = chave
        trocas += 1

    return data, comparacoes, trocas
```

```python
def bubble_sort_bioma(data):
    data = data.copy()
    comparacoes = 0
    trocas = 0
    n = len(data)

    for i in range(n):
        for j in range(0, n - i - 1):
            comparacoes += 1
            if data[j]['bioma'] > data[j + 1]['bioma']:
                data[j], data[j + 1] = data[j + 1], data[j]
                trocas += 1

    return data, comparacoes, trocas
```

```python
def quick_sort_bioma(data):
    comparacoes = [0]
    trocas = [0]

    def quick_sort(arr):
        if len(arr) <= 1:
            return arr
        else:
            pivot = arr[0]
            menores = []
            maiores = []
            for item in arr[1:]:
                comparacoes[0] += 1
                if item['bioma'] < pivot['bioma']:
                    menores.append(item)
                else:
                    maiores.append(item)
            trocas[0] += len(arr) - 1
            return quick_sort(menores) + [pivot] + quick_sort(maiores)

    resultado = quick_sort(data.copy())
    return resultado, comparacoes[0], trocas[0]
```

```python
def exibir_resultados(nome_algoritmo, dados_ordenados, comparacoes, trocas):
    print(f"\n{nome_algoritmo} - Resultados ordenados:")
    for linha in dados_ordenados[:10]:
        print(linha)
    print(f"Comparações realizadas: {comparacoes}")
    print(f"Trocas/Substituições realizadas: {trocas}")
```

```python
dados = df[['bioma', 'municipio']].dropna(subset=['bioma']).head(500).to_dict('records')

# Insertion Sort
ordenado, comp, troc = insertion_sort_bioma(dados)
exibir_resultados("Insertion Sort", ordenado, comp, troc)

# Bubble Sort
ordenado, comp, troc = bubble_sort_bioma(dados)
exibir_resultados("Bubble Sort", ordenado, comp, troc)

# Quick Sort
ordenado, comp, troc = quick_sort_bioma(dados)
exibir_resultados("Quick Sort", ordenado, comp, troc)
```
